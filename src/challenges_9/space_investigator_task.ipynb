Write a method `spaceInvestigator`, which would take a `String` as an argument, and which would check and fix the usage of spaces in a text, returning a fixed text as a `String`.

The rules for using spaces can be found here:
https://www.noslangues-ourlanguages.gc.ca/en/writing-tips-plus/punctuation-spacing

Check and fix as many rules, as you can, but it must be at least 6 different rules.
You may want to create a separate method for each step, which will be called from the method `spaceInvestigator`. Print a report on each step.
In your `main` method create an example `String`, which would contain all the errors you will be checking against.

Here are some of the rules, which you may want to check for:
- There should be no spaces before and after the text.
- There should never be no two spaces standing together, one after another.
- There must be a space after a punctuation mark (with some exceptions).
- There should not be no space before a punctuation mark (with some exceptions).
- There should be no space before or after a decimal period between numerals.
- There should be no space before or after a question or exclamation mark that is followed by a closing quotation mark, parenthesis or bracket.
- There should be no space before or after a comma that is followed by a closing quotation mark.
- There should be no space before or after dashes and hyphens when they are inserted between words, a word and a numeral, or two numerals.

## Step-by-Step Implementation Guide (Detailed)

1. Define the goal clearly.
   Build a method named `spaceInvestigator(String text)` that accepts a raw text and returns a corrected text where spacing rules are fixed.

2. Decide how your pipeline will work.
   Use a sequence of small helper methods. Each helper should fix one rule only. This makes your code easier to debug and explain.

3. Create the method skeletons first.
   Add `spaceInvestigator` and helper methods such as:
   - `trimOuterSpaces(String text)`
   - `removeMultipleSpaces(String text)`
   - `fixSpaceBeforePunctuation(String text)`
   - `fixSpaceAfterPunctuation(String text)`
   - `fixDecimalSpacing(String text)`
   - `fixSpacingBeforeClosingSymbols(String text)`
   - `fixCommaBeforeClosingQuote(String text)`
   - `fixHyphenAndDashSpacing(String text)`

4. Build a test input string in `main`.
   Write one long sample `String` that intentionally contains all spacing mistakes you want to fix. Include at least 6 different error types.

5. Implement Rule 1: leading and trailing spaces.
   In `trimOuterSpaces`, remove spaces before the first visible character and after the last visible character.

6. Implement Rule 2: repeated spaces.
   In `removeMultipleSpaces`, replace any sequence of 2 or more spaces with one single space.

7. Implement Rule 3: no space before punctuation.
   In `fixSpaceBeforePunctuation`, remove spaces before `. , ; : ! ?` (except where your chosen rule says otherwise).

8. Implement Rule 4: space after punctuation.
   In `fixSpaceAfterPunctuation`, ensure there is one space after punctuation when the next character starts a new word.

9. Implement Rule 5: decimal numbers.
   In `fixDecimalSpacing`, make sure decimal values like `12.34` have no spaces around the decimal point.

10. Implement Rule 6: punctuation before closing symbols.
    Remove spaces before `)` `]` `}` and before closing quotation marks after `?` or `!`, according to the selected style rules.

11. Implement Rule 7: comma before closing quotation marks.
    Ensure commas and closing quotes follow the selected spacing convention with no unwanted spaces.

12. Implement Rule 8: dashes and hyphens.
    Normalize spacing around hyphens/dashes when they connect words, word-number, or number-number combinations.

13. Chain all helper methods inside `spaceInvestigator`.
    Apply each helper one by one in a stable order. Store intermediate results in a variable and pass it to the next step.

14. Print a report for every step.
    After each helper call, print what rule was applied and show the updated text (or the number of changes). This is required by the task.

15. Compare original and final text in `main`.
    Print both versions so it is easy to verify what changed.

16. Verify at least 6 rules are actually working.
    Test with examples that trigger each rule. If a rule does not activate, update your input string so the report proves it works.

17. Keep methods focused and readable.
    One method = one responsibility. This improves maintainability and helps when your teacher checks the logic.

18. Optional final polish.
    Add comments, improve method names, and keep output format consistent so the result looks clean and professional.
