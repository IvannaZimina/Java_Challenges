# Bank account

Create a class Account for a bank account.

Create private fields for account characteristics like:
- account number.
- account balance.
- customer name.
- email.
- phone number.

Create getters and setters for each field.

Create two additional methods:
- one for depositing funds into the account.
- one for withdrawing funds from the account.

A customer should not be allowed to withdraw funds if that withdrawal takes their balance negative.

From your Main class with the usual main method create an instance of an Account class and then test your methods for withdrawing and depositing funds.

Print information to the console that confirms what the balance is after the methods are called.

Create an encapsulated class Account, where all fields are private. Set up getter and setter methods for all your fields.

The class should have two behavioral methods. One for depositing funds, the other for withdrawing funds.

In addition to this class, you need a Main class with a main method that creates at least one instance of the bank account class and simulates depositing and withdrawing money from the account.

## Step-by-step plan (what to do)

### 1. Create the `Account` class
Create a new Java class named `Account`.

### 2. Add private fields
Inside `Account`, add private fields:
- `accountNumber`
- `accountBalance`
- `customerName`
- `email`
- `phoneNumber`

### 3. Generate getters and setters
Create getter and setter methods for all fields.

### 4. Add `deposit` method
Create a method (for example `depositFunds(double amount)`) that:
- checks that amount is positive,
- adds amount to `accountBalance`,
- prints updated balance.

### 5. Add `withdraw` method
Create a method (for example `withdrawFunds(double amount)`) that:
- checks that amount is positive,
- checks that balance will not become negative,
- if enough money: subtracts amount and prints new balance,
- if not enough money: prints an error message and does not change balance.

### 6. Create `Main` class with `main`
In `Main.main(...)`:
- create one `Account` object,
- fill it with test data (number, name, email, etc.),
- call deposit and withdraw methods with different amounts.

### 7. Test scenarios
Run and check at least these cases:
- normal deposit,
- normal withdraw,
- withdraw more than balance,
- invalid values (0 or negative amount).

### 8. Verify task completion
Make sure:
- all fields are private,
- getters/setters exist,
- both behavior methods work,
- console output clearly shows the balance after operations.
